Scores by school spending identifies that quality and value are indicators of higher academic performance. The data shows that schools in the lower end of the spending spectrum had better passing rates than schools that spend more per student. 

The school size also shows that children perform better when there is more time dedicated to them; smaller class sizes help students perform better. Probably from low ratios between students and teachers while offering an environment where students can build their confidence socially and academically.

Comparing school types, charter schools do better academically since they need to maintain their public funding from governing bodies and philanthropies. This need for charter schools to fund possibly encourages proactive teaching and building upon strategies that helped their students succeed. 

Scores by school spending does not consider other factors for higher spending, such as extracurricular activates or school equipment/materials (laptops, workbooks, etc.) which district schools usually offer. Charter school families usually have the financial means to pay for extracurriculars (and more specialized extracurriculars) outside of the school day and school equipment/materials. More data is needed for the financial accessibility of extracurriculars and school equipment/materials for students between charter schools and district schools, and if possible, whether they are school or third-party funded. 

Although most charter schools are non-profit, they behave or are treated as for-profit and survive off public funds and philanthropic donations to stay open. Therefore, they have stricter standards to adhere by while also seeking approval for their efforts from philanthropic funding, usually by families that send their children to these schools. This limits the demographic diversity for charter schools due to affluent families doing whatever it takes (moving) to ensure their children get a private-school education at a public-school price but widens the inaccessibility of quality education for children and families that do that have the means to attend a charter school. Instead of encouraging charter schools and their lottery lists or school-choice, we should encourage the building of more district schools that lower the ratio between students and teachers which will naturally spread out the best teaching talent. This will discourage family migrations to high performing schools that force out underserved children and one day close the gap for accessibility of quality education.

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_names = school_data_complete["school_name"].unique()
school_total = len(school_names)
school_total
```

```python
# Calculate the total number of students
student_count = school_data_complete["student_name"]
total_student_count = len(student_count)
total_student_count
```

```python
# Calculate the total budget...first amount did not add correctly, so each school has its own set budget...
allocated_budgets = school_data_complete["budget"].unique()
total_budget = allocated_budgets.sum()
total_budget
```

```python
# Calculate the average (mean) math score
math_scores = school_data_complete["math_score"]
avg_math_scores = math_scores.mean()
avg_math_scores
```

```python
# Calculate the average (mean) reading score
reading_scores = school_data_complete["reading_score"]
average_reading_scores = reading_scores.mean()
average_reading_scores
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(total_student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(total_student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(total_student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({
    "Total Schools" : [school_total],
    "Total Students": [total_student_count],
    "Total Budget": [total_budget],
    "Average Math Score": [avg_math_scores],
    "Average Reading Score": [average_reading_scores],
    "% Math Passing": [passing_math_percentage],
    "% Reading Passing": [passing_reading_percentage],
    "% Overall Passing": [overall_passing_rate]
     })
# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

## School Summary

```python
# Use the code provided to select all of the school types
school_indexing_rmv = school_data_complete.drop(columns=["gender", "Student ID", "student_name", "grade", "reading_score", "math_score", "School ID", "size", "budget"])
#school_indexing_df = school_indexing.loc[:,["type"]]
school_indexing_df = school_indexing_rmv.drop_duplicates() #pandas library
school_indexing_df1 = school_indexing_df.reset_index() #pandas library
school_indexing = school_indexing_df1.drop(columns=["index"])
school_types = school_indexing.sort_values("school_name")
school_types
```

```python
# Calculate the total student count per school
per_school_counts = school_data_complete["school_name"].value_counts()
per_school_totals_df = pd.DataFrame(per_school_counts)
per_school_totals = per_school_totals_df.sort_values("school_name") #pandas library
per_school_totals
```

```python
# Calculate the total school budget and per capita (per student) spending per school
per_school_index_rmv = school_data_complete.drop(columns=["gender", "Student ID", "student_name", "grade", "reading_score", "math_score", "School ID", "size", "type"])
per_school_indexing_df = per_school_index_rmv.drop_duplicates() #pandas library
per_school_indexing_df1 = per_school_indexing_df.reset_index() #pandas library
per_school_indexing = per_school_indexing_df1.drop(columns=["index"])
per_school_budget = per_school_indexing.sort_values("school_name")
school_budget_stdntCount = pd.merge(per_school_budget, per_school_totals, how="right", on=["school_name"])
per_school_capita = school_budget_stdntCount["budget"]/school_budget_stdntCount["count"]
school_budget_stdntCount["Per Capita"] = per_school_capita
school_budget_stdntCount
```

```python
# Calculate the average test scores per school
#per_school_math = school_data_complete.set_index("school_name")
per_school_math_rmv = school_data_complete.drop(columns=["gender", "Student ID", "student_name", "grade", "reading_score" ,"School ID", "size", "type", "budget"])
per_school_math_sum = per_school_math_rmv.groupby(["school_name"]).sum()
#per_school_reading =
per_school_read_rmv = school_data_complete.drop(columns=["gender", "Student ID", "student_name", "grade", "math_score", "School ID", "size", "type", "budget"])
per_school_read_sum = per_school_read_rmv.groupby(["school_name"]).sum()
per_school_mtrd = pd.merge(per_school_math_sum, per_school_read_sum, how="right", on=["school_name"])
semi_per_schools = pd.merge(per_school_mtrd, school_budget_stdntCount, how="right", on=["school_name"])
semi2_per_schools = pd.merge(semi_per_schools, school_types, how="right", on=["school_name"])
avg_math = semi2_per_schools["math_score"]/semi2_per_schools["count"]
avg_read = semi2_per_schools["reading_score"]/semi2_per_schools["count"]
semi2_per_schools["Average Math Score"] = avg_math
semi2_per_schools["Average Reading Score"] = avg_read
semi3_per_schools = semi2_per_schools.drop(columns=["math_score", "reading_score"])
semi3_per_schools
```

```python
# Calculate the number of students per school with math scores of 70 or higher
per_school_math_rmv1 = school_data_complete.drop(columns=["gender", "Student ID", "student_name", "budget", "grade", "reading_score", "School ID", "size", "type"])
passing_per_school_math = per_school_math_rmv1[(school_data_complete["math_score"] >= 70)]
per_school_math_count = passing_per_school_math.groupby(["school_name"]).count()
semi4_per_schools = pd.merge(semi3_per_schools, per_school_math_count, how="right", on=["school_name"])
per_school_math_count
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
per_school_read_rmv1 = school_data_complete.drop(columns=["gender", "Student ID", "student_name", "budget", "grade", "math_score", "School ID", "size", "type"])
passing_per_school_read = per_school_read_rmv1[(school_data_complete["reading_score"] >= 70)]
per_school_read_count = passing_per_school_read.groupby(["school_name"]).count()
semi5_per_schools = pd.merge(semi4_per_schools, per_school_read_count, how="right", on=["school_name"])
per_school_read_count
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
per_school_both_rmv = school_data_complete.drop(columns=["gender", "Student ID", "student_name", "budget", "grade", "School ID", "size", "type"])
pass_per_school_both = per_school_both_rmv[(school_data_complete["math_score"] >= 70) &(school_data_complete["reading_score"] >= 70)]
per_school_both = pass_per_school_both.groupby(["school_name"]).count()
per_school_both_rnm = per_school_both.rename(columns={"reading_score": "Passed Math & Reading"})
per_school_both_df = per_school_both_rnm.drop(columns=["math_score"])
semi6_per_schools = pd.merge(semi5_per_schools, per_school_both_df, how="right", on=["school_name"])
per_school_both_df
```

```python
# Use the provided code to calculate the passing rates
per_school_pass_math = semi6_per_schools["math_score"] / semi6_per_schools["count"] * 100
per_school_pass_reading = semi6_per_schools["reading_score"] / semi6_per_schools["count"] * 100
overall_pass_rate = semi6_per_schools["Passed Math & Reading"] / semi6_per_schools["count"] * 100
semi6_per_schools["Passing Math Rate"] = per_school_pass_math
semi6_per_schools["Passing Reading Rate"] = per_school_pass_reading
semi6_per_schools["Overall Passing (Math & Reading) Rate"] = overall_pass_rate
semi7_per_schools = semi6_per_schools.rename(columns={"reading_score": "Passed Reading", "Per Capita": "Per Student Budget", "math_score": "Passed Math", "type":"School Type", "count":"Student Count", "budget":"Total School Budget","school_name": "School Name"})
semi7_per_schools
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = semi7_per_schools

# Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
per_school_summary
```

## Math Scores by Grade

```python
# Use the code provided to separate the data by grade
school_grades_math = school_data_complete.drop(columns=["gender", "Student ID", "student_name", "reading_score", "type", "School ID", "size", "budget"])

ninth_graders_math = school_grades_math.loc[(school_grades_math["grade"] == "9th")]
tenth_graders_math = school_grades_math.loc[(school_grades_math["grade"] == "10th")]
eleventh_graders_math = school_grades_math.loc[(school_grades_math["grade"] == "11th")]
twelfth_graders_math = school_grades_math.loc[(school_grades_math["grade"] == "12th")]

# Group by `school_name` and take the mean of the `math_score` column for each.
ninth_grade_math_scores = ninth_graders_math.groupby(["school_name", "grade"]).mean()
tenth_grade_math_scores = tenth_graders_math.groupby(["school_name", "grade"]).mean()
eleventh_grade_math_scores = eleventh_graders_math.groupby(["school_name", "grade"]).mean()
twelfth_grade_math_scores = twelfth_graders_math.groupby(["school_name", "grade"]).mean()

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.merge(ninth_grade_math_scores, tenth_grade_math_scores, how="right", on=["school_name"])
math_scores_by_grade = pd.merge(math_scores_by_grade, eleventh_grade_math_scores, how="right", on=["school_name"])
math_scores_by_grade = math_scores_by_grade.rename(columns={"math_score_x": "9th"})
math_scores_by_grade = math_scores_by_grade.rename(columns={"math_score_y": "10th"})
math_scores_by_grade = math_scores_by_grade.rename(columns={"math_score": "11th"})
math_scores_by_grade = pd.merge(math_scores_by_grade, twelfth_grade_math_scores, how="right", on=["school_name"])
math_scores_by_grade = math_scores_by_grade.rename(columns={"math_score": "12th"})

# Minor data wrangling
math_scores_by_grade.index.name = None

# Display the DataFrame
math_scores_by_grade
```

## Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.
school_size_df = per_school_summary.copy()
school_size_df["School Size"] = pd.cut(school_size_df["Student Count"], size_bins, labels=size_labels, right=False)#function needed, set to false
school_size_df
```

```python
# Calculate averages for the desired columns. 
size_math_scores = school_size_df.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = school_size_df.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = school_size_df.groupby(["School Size"])["Passing Math Rate"].mean()
size_passing_reading = school_size_df.groupby(["School Size"])["Passing Reading Rate"].mean()
size_overall_passing = school_size_df.groupby(["School Size"])["Overall Passing (Math & Reading) Rate"].mean()
```
